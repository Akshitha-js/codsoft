```python
Fraud_data1 = Fraud_data.drop(['Unnamed: 0', 'trans_date_trans_time', 'cc_num', 'merchant', 'category',
                               'first', 'last', 'gender', 'street', 'city', 'state', 'city_pop','dob',
                               'trans_num','unix_time','merch_lat', 'merch_long'],axis=1)
```

```python
Fraud_data1.job.value_counts()
Fraud_data1.job=pd.Categorical(Fraud_data1.job).codes
Fraud_data1
```

```python
# Defining the Features/Independent Variables (X) and Target variable/Dependent Variable (y) 
X_data = Fraud_data1.drop('is_fraud', axis=1)
y_target = Fraud_data1['is_fraud']
```

```python
# Importing necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Splitting the dataset 
X_train,X,Y_train,Y = train_test_split(X_data,y_target,test_size = 0.3,random_state=2)
```

```python
X_Val,X_test,Y_Val,Y_test = train_test_split(X,Y,test_size = 0.5,random_state=2)
```

```python
# Model 1: Logistic Regression
logistic_model = LogisticRegression()
logistic_model=logistic_model.fit(X_train, Y_train)
logistic_predictions = logistic_model.predict(X_test)
```

```python
# Model 2: Decision Tree
tree_model = DecisionTreeClassifier(max_depth = 5)
tree_model=tree_model.fit(X_train, Y_train)
tree_predictions = tree_model.predict(X_test)
```

```python
# Model 3: Random Forest
forest_model = RandomForestClassifier(n_estimators=50,max_depth = 5,random_state = 2)
forest_model=forest_model.fit(X_train, Y_train)
forest_predictions = forest_model.predict(X_test)
```

```python

name = {
    "LR": logistic_model,
    "DTC":tree_model,
    "RFC":forest_model ,   
}
```

```python
#Evaluation
def evaluate_model(name, predictions):
    print(f"------- {name} -------")
    print("Accuracy:", accuracy_score(Y_test, predictions))
    print("Confusion Matrix:\n", confusion_matrix(Y_test, predictions))
    print("Classification Report:\n", classification_report(Y_test, predictions))
```
